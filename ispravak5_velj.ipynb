```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/bio-chem-info/python_course/refs/heads/main/preprocess_data.csv')
```

```python
X = dataset.iloc[:,:-1] #ne ukljuci zadnju kolonu
Y = dataset.iloc[:,-1] #ukljuci samo zadnju kolonu, to je vektor koji zelimo predvidjeti
```

```python
X_encoded = pd.get_dummies(X, columns=['Country'], drop_first=True)
X_encoded
X_encoded = X_encoded.replace({True:1,False:0})
Y_encoded = Y.replace({'Yes':1,'No':0})
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, Y_encoded, test_size =0.25, random_state = 42)
```

#skaliranje

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc
```

```python
sc.fit(X_train)
```

```python
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train_sc = pd.DataFrame(X_train_sc, columns=X_train.columns, index = X_train.index)
```
