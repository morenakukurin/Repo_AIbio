```python
def check_gender_by_name(name):
  if name[-1] == 'a': #za definirati zadje slovo imena
    gender = 'female'
  elif name[-1]=='o':
    gender = 'male'
  else:
    gender = 'unknown'
  return gender
```

```python
def ph(c):
  return -math.log10(c)
```

```python
def ph_evaluator(c=0.001):
  '''pH decision maker.'''
  ph_value = ph(c)
  if ph_value < 7:
    decision = 'acid'
  else:
    decision = 'luznato'
  return decision
```

```python
ph_evaluator(0.002)
```

```python
def ph_evaluator(c=0.001):
  if ph(c) < 7:
    return "kiselo"
  else:
    return 'luznato'
conc_list = [0.001, 0.01, 1, 10]

for solution in conc_list:
  print(ph_evaluator(solution))
```
