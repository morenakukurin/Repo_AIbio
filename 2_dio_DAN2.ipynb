```python
set_versi = iris_df[iris_df['species'].isin(['setosa','versicolor'])]
```

```python
set_versi
```

```python
#postoji li stat znac razlika izmedu dvije grupe - t test

grupa1 = set_versi[set_versi['species']=='setosa']['sepal length (cm)']
```

```python
grupa1
```

```python
grupa2 = set_versi[set_versi['species']=='versicolor']['sepal length (cm)']
```

```python
# t test
ttest = pg.ttest(grupa1,grupa2)
print(f't test rezultati:\nf{ttest}') #p value <0.05 postoji stat znac razlika
```

```python
sirina = pg.ttest(iris_df[iris_df['species']=='setosa']['sepal length (cm)'], iris_df['species']=='versicolor'['sepal length (cm)'])
```

```python
#anova
anova = pg.anova(dv='sepal length (cm)', between = 'species', data = iris_df)
```
